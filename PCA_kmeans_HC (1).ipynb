```python
var_type = 'phylum'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
df = pd.read_csv('generaABModels.csv')
```

```python
df1=df[['poscount',	'negcount',	'posper',	'negper',	'epiratio']]
```

```python
scaler=StandardScaler()
df_scaled=scaler.fit_transform(df1)
```

```python
pca=PCA(n_components=3)
pca_df=pca.fit_transform(df_scaled)
```

```python
plot_df = pd.DataFrame()
plot_df['pc1']=pca_df[:,0]
plot_df['pc2']=pca_df[:,1]
```

```python
plot_df[var_type]=df[var_type]
```

```python
sns.scatterplot(data=plot_df,x='pc1', y='pc2',hue=var_type)
plt.savefig(var_type+'pca.png',bbox_inches='tight')
```
